```python
import soundfile as sndfl
import numpy as np
import matplotlib.pyplot as plt
```

```python
speech, fs1 = sndfl.read("speech.wav")
noise, fs2 = sndfl.read("noise.wav")
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout = 2)
fig.suptitle("Signal as function of time")
t = np.arange(0, len(speech)/fs1, 1/fs1)
axs[0].plot(t, speech)
axs[0].set_title('Speech Signal')
axs[0].set_xlabel('time')
axs[0].set_ylabel('amplitude')

t2 = np.arange(0, len(noise)/fs2, 1/fs2)
axs[1].plot(t2, noise)
axs[1].set_title('Noise Signal')
axs[1].set_xlabel('time')
axs[1].set_ylabel('amplitude')

plt.show()
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout = 2)
fig.suptitle("Instantaneous Power of Signal")
t = np.arange(0, len(speech)/fs1, 1/fs1)
axs[0].plot(t, np.square(speech))
axs[0].set_title('Speech Signal')
axs[0].set_xlabel('time')
axs[0].set_ylabel('Power')

t2 = np.arange(0, len(noise)/fs2, 1/fs2)
axs[1].plot(t2, np.square(noise))
axs[1].set_title('Noise Signal')
axs[1].set_xlabel('time')
axs[1].set_ylabel('Power')

plt.show()
```

```python
total_energy_speech = np.sum(np.square(speech))
print(total_energy_speech)
```

```python
total_energy_noise = np.sum(np.square(noise))
print(total_energy_noise)
```

```python
print(10*np.log10(total_energy_speech/total_energy_noise))
```

Noise signal is ~15db higher than the speech signal
