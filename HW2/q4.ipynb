```python
import numpy as np
import soundfile as sndfl
import matplotlib.pyplot as plt
```

```python
signal, fs = sndfl.read("speech2.wav")
```

```python
total_energy_speech = np.sum(np.square(signal))
print(total_energy_speech)
```

```python
x = np.asarray(signal, dtype=float)
N = x.shape[0]
n = np.arange(N)
k = n.reshape((N, 1))
M = np.exp(-2j * np.pi * k * n / N)
X = np.dot(M,x)
```

```python
fig, axs = plt.subplots(2, 1, constrained_layout = 2)
t = np.arange(0, len(signal)/fs, 1/fs)
axs[0].plot(t, signal)
axs[0].set_title('Time Domain')
axs[0].set_xlabel('time (in s)')
axs[0].set_ylabel('amplitude')

axs[1].plot(np.abs(X))
axs[1].set_title('Magnitude Response')
axs[1].set_xlabel('frequency')
axs[1].set_ylabel('magnitude')

plt.show()
```

```python
np.sum(np.square(np.abs(X)))/len(signal)
```

Yes the energy computed in the time domain is equal to the energy that is computed in the frequency domain. 
Therefore, Parseval's Theorem holds true in practice
